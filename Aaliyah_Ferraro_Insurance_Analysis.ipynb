```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('insurance.csv')

# Dataset source: https://www.kaggle.com/datasets/willianoliveiragibin/healthcare-insurance?resource=download
```

# #Question 1 (Easy): What is the average medical insurance charge of individuals in the dataset?

```python
average_charge = df['charges'].mean()
print(f'Average Medical Insurance Charge: ${average_charge:.2f}')
```

The variable average_charge extracts the 'charges' column from the insurance dataset and calculates the average amount of insurance fees for each observation using the .mean() function. The output is a floating-point number that represents the general amount that the dataset's sample population is expected to pay.

The result, an average amount of $13,270.42, directly answers the question using statistical mean. This value indicates that the people included in this study generally face about $13k in annual insurance charges.


#Question 2 (Medium): How do insurance charges differ between smokers and non-smokers?

```python
charges_by_smoker = df.groupby('smoker')['charges'].mean()
for status, charge in charges_by_smoker.items():
  print(f'Average Insurance Charges for {status}: ${charge:.2f}')

charges_by_smoker.plot(kind='bar', color=['darkgreen','maroon'], title='Average Insurance Cost by Smoking Status')
plt.xlabel('Smoker')
plt.ylabel('Average Charges')
```

The variable charges_by_smoker groups the columns 'smoker' and 'charges' together, wherein the 'smoker' column, a boolean-like variable with yes and no, is split into two groups and is the x-value, and the 'charges' column is the insurance charges and the y-value. The .mean() function calculates the average insurance charges for smokers and non-smokers, which can be seen through the bar chart.

The results reveal a large disparity between the insurance charges of smokers and non-smokers. The insurance cost of smokers, at \$32,050,23, is greater than that of non-smokers by three times, at \$8,434.27. This difference shows that insurance companies account for their client's health status when calculating prices. Since smoking is associated with long-term health complications, insurance companies charge higher, which highlights the use of behavioral risk in insurance pricing.

#Question 3 (Medium): Are there differences in average insurance charges depending on the region people live in?

```python
charges_by_region = df.groupby('region')['charges'].mean()
print(charges_by_region)

charges_by_region.plot(kind='bar', title='Average Charges by Region', color='lightblue')
plt.xlabel('Region')
plt.ylabel('Average Charges')
```

The variable charges_by_region groups the columns 'region' and 'charges' together, wherein the 'region' column, a variable that is split into four subgroups and is the x-value, and the 'charges' column is the insurance charges and the y-value. The .mean() function calculates the average insurance charges for the regions northeast, northwest, southeast, and southwest, which can be seen through the bar chart.

The results show a relatively consistent insurance cost across regions with the difference ranging between \$1k to \$2k. The average cost for the Northeast is \$13,406.38, \$12,427.57 for the Northwest, \$14,735.41 for the Southeast, and \$12,346.94 for the Southwest. The Southeast has the highest average and Southwest with the lowest, but the small differences between each region suggest that location is not a strong determining factor in insurance costs. The statistics and bar chart demonstrate that compared to other variables, such as smoking status and age, geography has a less significant role in insurance pricing.


#Question 4 (Hard): How do age, smoking, and BMI affect insurance costs?

```python
colors = {
    'yes' : 'crimson',
    'no' : 'lightsteelblue'
    }

plt.figure(figsize=(20,15))
plt.scatter(df['age'], df['charges'],
            color=df['smoker'].map(colors),
            s=df['bmi'], alpha=0.5)

plt.title('Charges by Age, BMI, and Smoking')
plt.xlabel('Age')
plt.ylabel('Charges')
```

To find how age, smoking status, and BMI collectively affect insurance costs, a dictionary called 'colors' is created and contains the keys 'yes' and 'no' to categorize smoking status with the values 'crimson' and 'lightsteelblue,' respectively. Then, a scatterplot is used to represent an individual's age (x-axis) and insurance charge (y-axis). Each dot, according to the dictionary, shows whether they smoke or not; and the size of each represents their BMI, so the higher the BMI, the larger a dot.

The plot shows a trend wherein smokers have a significantly higher insurance cost than non-smokers, especially as age increases. Furthermore, higher BMI values appear where insurance cost is higher, showing a positive correlation with smoking status, age, and insurance cost. This suggests that these factors have a strong collective impact on the amount people pay for healthcare, demonstrating how lifestyle and phsyical health, with factors such as smoking and obesity, contribute to increased medical risks, and, consequently, insurance charges.
